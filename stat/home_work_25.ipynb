```python
import math


n = 64
x_bar = 84.0
s = 24.0
alpha = 0.10


se = s / math.sqrt(n)


try:
    from scipy.stats import t
    df = n - 1
    t_crit = t.ppf(1 - alpha/2, df)
except Exception:
   
    t_crit = 1.645

me = t_crit * se
ci_lower = x_bar - me
ci_upper = x_bar + me

print("t критическое =", t_crit)
print(f"90% CI = ({ci_lower:.2f}, {ci_upper:.2f})")
```
