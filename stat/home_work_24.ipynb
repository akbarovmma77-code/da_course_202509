```python
from scipy.stats import norm
```

№1

```python
z = 1.27
probality = norm.cdf(z)
print(probality) 
```

№2

```python
z1 = -1.76
z2 = -1.19
probality = norm.cdf(z2) - norm.cdf(z1)
print(probality)
```

```python
mu = 100
sigma = 15
x = 140

z  = (x-mu)/sigma

probality = 1-norm.cdf(z)
print(probality)
```
