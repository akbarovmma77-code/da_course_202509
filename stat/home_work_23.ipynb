```python
import math as mt
```

№1

```python
a = 4
k = 2
outcomes_t = 2**a
outcomes_f = mt.factorial(a)/(mt.factorial(k)*mt.factorial(a-k))
print(outcomes_f/outcomes_t) #вероятнсть что орел выпал два раза
```

№2

```python
a = 4
not_k = 1
outcomes_t  = 2**a
k_false = not_k/outcomes_t
k_true  = 1-k_false
print(k_true) #вероятность что выпало 1 или больше орлов

```
