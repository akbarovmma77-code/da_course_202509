```python
w = 3
z = 4
r = 5
wzr = w+z+r
C = mt.factorial(wzr)/(mt.factorial(2)*mt.factorial(wzr-2))
zr  = z*r
result = zr/C
print(result*100,"%",sep="") #вероятность того что я возьму красный и зеленый
```
